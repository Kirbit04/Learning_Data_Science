Importing Requests for requesting the API

```python
import requests
```

Setting the base url

```python
base_url = "https://pokeapi.co/api/v2/"
```

Defining requests, accessing API and exploring data in the API

```python
def get_pokemon_info(id):
  url = f"{base_url}/pokemon/{id}"
  response = requests.get(url)
  if response.status_code == 200:
    pokemon_data = response.json()
    return pokemon_data
  else:
    print(f"Error: {response.status_code}")

all_pokemon_data = []

for i in range(1, 1025 + 1):
  pokemon_id = i
  pokemon_info = get_pokemon_info(pokemon_id)
  if pokemon_info:
    pokemon_data_list ={
      'ID': pokemon_info.get('id'),
      'Name': pokemon_info.get('name').capitalize(),
      'Height': pokemon_info.get('height'),
      'Weight': pokemon_info.get('weight'),
      'Base_experience': pokemon_info.get('base_experience'),
      'Abilities': ', '.join(ability['ability']['name'] for ability in pokemon_info.get('abilities', [])),
      'Types': ', '.join(type['type']['name'] for type in pokemon_info.get('types', [])),
      'Battle_items': ','.join(held_items['item']['name']for held_items in pokemon_info.get('held_items', [])),
      'Moves': ', '.join(move['move']['name'] for move in pokemon_info.get('moves',[])),
      'Stats': ', '.join(stat['stat']['name'] for stat in pokemon_info.get('stats',[]))
      }
    all_pokemon_data.append(pokemon_data_list)
    print(f"Processed {pokemon_info['name'].capitalize()}(ID: {pokemon_id})")
  else:
    print(f"Pokemon with ID {pokemon_id} not found.")
```

```python
import csv
#data fetching and storing in csv
csv_file_path = "pokemon_data.csv"
#column headers of csv
csv_headers = ["ID", "Name", "Height", "Weight", "Base_experience", "Abilities", "Types", "Battle_items", "Moves", "Stats"]

#writing data to csv
try:
  with open(csv_file_path, "w", newline="") as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=csv_headers)
    #write header row
    csv_writer.writeheader()
    #write data rows
    csv_writer.writerows(all_pokemon_data)
  print(f"Data has been successfully written to {csv_file_path}")
except Exception as e:
  print(f"Error writing to CSV: {e}")
```
